```python
import pandas as pd
```

```python
df = pd.read_csv('human_disease_integrated_full.tsv', sep='\t',
                names = ['Gene_Identifier','Gene_Name','DOID','Disease','Unnamed'])
```

```python
df.head()
```

```python
df
```

```python
asthma_group = df.groupby(df.DOID)
asthma_group1 = asthma_group.get_group("ICD10:J45")
asthma_group1
```

```python
asthma_group2 = df.groupby(df.DOID)
asthma_group3 = asthma_group2.get_group("DOID:2841")
asthma_group3
```

```python
obesity_group = df.groupby(df.DOID)
obesity_group1 = obesity_group.get_group("DOID:9970")
obesity_group1
```

```python
obesity_group2 = df.groupby(df.DOID)
obesity_group3 = obesity_group2.get_group("ICD10:E66")
obesity_group3
```

```python
asthma = [asthma_group1,asthma_group3]
total_asthma = pd.concat(asthma)
```

```python
print(asthma)
```

```python
obesity = [obesity_group1, obesity_group3]
total_obesity = pd.concat(obesity)
print(total_obesity)
```

```python
print(total_asthma)
```

```python
total_asthma
```

```python
total_obesity
```

```python
match_df = total_asthma['Gene_Name'].to_list()
output = []
for Gene_Name in total_obesity['Gene_Name']:
    if Gene_Name in match_df:
        output.append(Gene_Name)
```

```python
print('\n'.join(map(str, output)))
```
